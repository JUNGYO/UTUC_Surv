```python
import pandas as pd
import numpy as np
import os
```

```python
datapath = './data'
model_path = './save/'

## data
# AMC = pd.read_csv(os.path.join(datapath+'/'+'UTUC_deep.csv'))
ext_Hallym = pd.read_csv(os.path.join(datapath+'/'+'UTUC_deep_Hallym.csv'))
ext_SNU = pd.read_csv(os.path.join(datapath+'/'+'UTUC_deep_SNU.csv'))

train_data = pd.read_csv(os.path.join(model_path+'/'+'train_data.csv'))
test_data = pd.read_csv(os.path.join(model_path+'/'+'test_data.csv'))
valid_data = pd.read_csv(os.path.join(model_path+'/'+'valid_data.csv'))
```

```python
AMC = pd.concat([train_data,test_data,valid_data],axis=0)
```

```python
variables=['Age', 'Female', 'BMI', 'Pre_GFR','Location_Pelvis', 'Location_multiple', 'HUN', 'path_T', 'Path_N','Path_Grade',"Path_CIS","Prg","PFS","OS","Surv"]
```

```python
AMC = AMC[variables]
ext_Hallym = ext_Hallym[variables]
ext_SNU  = ext_SNU[variables]
```

```python
AMC['hospital'] = 'AMC'
ext_Hallym['hospital'] = 'Hallym'
ext_SNU['hospital'] = 'SNU'
```

```python
AMC['Path_N'].replace({' ':np.nan},inplace=True)
AMC['Path_N'].replace({2:1},inplace=True)
AMC['Path_N'] = AMC['Path_N'].astype(float)

AMC['Location_multiple'].replace({1:0},inplace=True)
AMC['Location_multiple'].replace({2:1},inplace=True)
AMC['Location_multiple'].replace({3:1},inplace=True)
AMC['Location_multiple'].replace({4:1},inplace=True)
```

```python
merge = pd.concat([AMC, ext_Hallym, ext_SNU], ignore_index=True)
```

```python
from tableone import TableOne
```

```python
AMC.columns
```

```python
AMC = AMC.reset_index()
```

```python
nonnormal= ['Age', 'BMI', 'Pre_GFR', 'PFS', 'Surv']
categorical = ['Female', 'Location_Pelvis', 'Location_multiple', 'HUN', 'path_T', 'Path_N', 'Path_Grade',
               'Path_CIS', 'Prg', 'OS','hospital']
```

```python
Table1= TableOne(AMC, nonnormal=nonnormal, categorical=categorical, dip_test=True, normal_test=True, tukey_test=True)
```

```python
Table1
```

```python
Table2= TableOne(merge,nonnormal=nonnormal, categorical=categorical, dip_test=True, normal_test=True, tukey_test=True)
```

```python
Table2
```

```python
Table3= TableOne(merge, groupby=['hospital'], nonnormal=nonnormal, categorical=categorical, dip_test=True, normal_test=True, tukey_test=True, pval=True, htest_name=True)
```

```python
Table1.to_csv(datapath+'/'+'Table1.csv')
Table2.to_csv(datapath+'/'+'Table2.csv')
Table3.to_csv(datapath+'/'+'Table3.csv')
```
